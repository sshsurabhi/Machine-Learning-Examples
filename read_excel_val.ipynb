```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import os

#fig, ax = plt.subplots()

csv_directory= "C:/Users/surab/OneDrive/Desktop/csvfiles/"

files_in_directory= os.listdir(csv_directory)

x_all =[]
y_all=[]

for i in range(len(files_in_directory)):
    X=[]
    Y=[]

    with open(csv_directory+files_in_directory[i],'r') as csvfile:
        reader= csv.reader(csvfile, delimiter=',')
        for row in  reader:
            x1=float(row[0])
            y1=float(row[1])
            X.append(x1)
            Y.append(y1)
            
        a=np.array(X).reshape(76,-1)
        #print(a)
        b=np.array(Y).reshape(76,-1)
        c=np.mean(a,axis=1)
        d=np.mean(b,axis=1)
        #print(c.shape,d.shape)
        #print("hiii",c)
        x_all.append(c)
        y_all.append(d)
        
for i in range(len(x_all)):
    #plt.plot(x_all[i], y_all[i])
    lines = plt.plot(x_all[i],y_all[i])
    plt.legend(lines)
#plt.legend()
plt.show()


plt.savefig('out.png')
    
```
